```python
def twint_time(start, end, user):
    # Input three params start and end dates in the format '20xx-mm-dd' and a user such as 'LexFridman'
    # All params should be string
    # There should be no spaces in the username
    c = twint.Config()
    c.Username = user
    c.Since = start
    c.Until = end
    # Custom output format
    c.Limit = 10
    c.Pandas = True
    c.Pandas_clean = True
    c.Store_pandas = True
    c.Hide_output = True
    twint.run.Search(c)
    df.df = twint.storage.panda.Tweets_df
    return df.df.head()
```
